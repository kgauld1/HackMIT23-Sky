```python
import json
import time
import pickle
from copy import deepcopy
import matplotlib.pyplot as plt
from datetime import datetime 
%matplotlib notebook
```

```python
jsonfn = 'dawndata.json'
with open(jsonfn, 'r') as datafile:
    data = json.load(datafile)
    print(len(data))
    print('done loading!')
```

```python
with open('jsdata.pkl', 'rb') as readfile:
    jsdata = pickle.load(readfile)
```

```python
rft_data = deepcopy(jsdata)
for x in rft_data:
    x['time'] = datetime.timestamp(datetime.strptime(x['time'], '%Y-%m-%dT%H:%M:%S.000Z'))
with open('timefix.pkl', 'wb') as f:
    pickle.dump(rft_data, f)
```

```python
pdata = [x for x in jsdata if x['user_id'] == 593]
```

```python
plt.figure()
px = [x['position_y'] for x in pdata]
pt = [x['time'] for x in pdata]
plt.plot(pt, px)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot([x['position_x'] for x in pdata],
        [x['position_y'] for x in pdata],
        [x['position_z'] for x in pdata])


ax.set_xlabel('$X$')
ax.set_ylabel('$Y$')
ax.set_zlabel('$Z$')
```
